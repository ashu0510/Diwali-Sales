***#Exploratory Data Analysis***

Gender

```python
df.columns
```

```python
sns.countplot(x='Gender', data=df)
```

```python

# Define a custom color palette
custom_palette = ["#FF5733", "#33FF57"]  # Example colors, you can specify your own

# Create the countplot with the custom palette
ax = sns.countplot(x='Gender', hue='Gender', data=df, palette=custom_palette, legend=True)

# Add labels to the bars
for bars in ax.containers:
    ax.bar_label(bars)

# Show the plot
plt.show()
```

```python
#Using Groupby
df.groupby(['Gender'], as_index = False)['Amount'].sum().sort_values(by = "Amount", ascending = False)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Grouping data by Gender and calculating total sales amount
sales_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by="Amount", ascending=False)

# Define a custom color palette
custom_palette = ["#FF5733", "#33FF57"]  # Example colors, you can specify your own

# Create the barplot with custom colors using hue
sns.barplot(x='Gender', y='Amount', hue='Gender', data=sales_gen, palette="husl", legend=False)

# Add labels and title
plt.xlabel('Gender')
plt.ylabel('Total Sales Amount')
plt.title('Total Sales Amount by Gender')

# Show the plot
plt.show()
```

From above graph we can see that most of the buyers are females and even the purchasing power of the females are greater than male.


**Age**

```python
# Define a custom color palette
custom_palette = ["#FF5733", "#33FF57"]  # Example colors, you can specify your own

# Create the countplot with the custom palette
ax = sns.countplot(x='Age Group', hue='Gender', data=df, palette=custom_palette, legend=True)

# Add labels to the bars
for bars in ax.containers:
    ax.bar_label(bars)

# Show the plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Grouping data by Age Group and Gender, and calculating total sales amount
sales_age_gender = df.groupby(['Age Group', 'Gender'], as_index=False)['Amount'].sum().sort_values(by="Amount", ascending=False)

# Create the barplot with Seaborn's default palette
sns.barplot(x='Age Group', y='Amount', hue='Gender', data=sales_age_gender)

# Add labels and title
plt.xlabel('Age Group')
plt.ylabel('Total Sales Amount')
plt.title('Total Sales Amount by Age Group and Gender')

# Show the plot
plt.show()
```

From above graph we can see that most of the buyers are of age group between 26-35 females

**State**

```python
df.columns
```

```python
#Total number of orders from top 10 states
sales_state= df.groupby(['State'], as_index= False)['Orders'].sum().sort_values(by= 'Orders', ascending= False).head(10)

sns.set(rc= {'figure.figsize':(15,5)})
sns.barplot(data= sales_state, x= 'State', y= 'Orders')
```

```python
#Total amount/sales from top 10 states
sales_state= df.groupby(['State'], as_index= False)['Amount'].sum().sort_values(by= 'Amount', ascending= False).head(10)

sns.set(rc= {'figure.figsize':(15,5)})
sns.barplot(data= sales_state, x= 'State', hue='State', y= 'Amount', palette='viridis')
```

From above graph we can see that most of the orders & totalsales/amount are from UP, Maharastra and Karnataka respectively.

**Marital Status**

```python
ax = sns.countplot(data= df, x= 'Marital_Status',hue= 'Marital_Status', palette='rocket')

sns.set(rc= {'figure.figsize':(7,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state= df.groupby(['Marital_Status', 'Gender'], as_index= False)['Amount'].sum().sort_values(by= 'Amount', ascending= False).head(10)

sns.set(rc= {'figure.figsize':(6,5)})
sns.barplot(data= sales_state, x= 'Marital_Status', hue='Gender', y= 'Amount', palette='rocket')
```

From above graph we can see that most of the buyers are married(women) and they have high purchasing power.

**Occupation**

```python
ax = sns.countplot(data= df, x= 'Occupation', hue= 'Occupation', palette='Paired')

sns.set(rc= {'figure.figsize':(20,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state= df.groupby(['Occupation'], as_index= False)['Amount'].sum().sort_values(by= 'Amount', ascending= False).head(10)

sns.set(rc= {'figure.figsize':(20,5)})
sns.barplot(data= sales_state, x= 'Occupation', hue='Occupation', y= 'Amount', palette='Paired')
```

From the above graph we can see that most of the buyers are working in IT, Aviation and Healthcare sector.

**Product Category**

```python
ax = sns.countplot(data=df, x='Product_Category', hue='Product_Category', palette='Paired')

sns.set(rc={'figure.figsize': (20, 15)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state= df.groupby(['Product_Category'], as_index= False)['Amount'].sum().sort_values(by= 'Amount', ascending= False).head(10)

sns.set(rc= {'figure.figsize':(20,5)})
sns.barplot(data= sales_state, x= 'Product_Category', hue='Product_Category', y= 'Amount', palette='Paired')
```

From the above graph we can see that most of the sold products are from are food, clothing and electronics.

```python
sales_state= df.groupby(['Product_ID'], as_index= False)['Amount'].sum().sort_values(by= 'Amount', ascending= False).head(10)

sns.set(rc= {'figure.figsize':(20,5)})
sns.barplot(data= sales_state, x= 'Product_ID', hue='Product_ID', y= 'Amount', palette='Paired')
```
